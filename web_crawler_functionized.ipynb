```python
def rating_crawler(time_id):
    show = requests.get('https://www.xxx.com.tr/xxx-xxx'+'?tarih='+ time_id)
    source = BeautifulSoup(show.content,"html.parser")
    
    main = source.find("div",attrs={"class":"find your class to be crawled and write here"})
    for_ratings = main.find("div",attrs={"class":"clearfix row"})
    
    n_columns = 0
    n_rows=0
    column_names = []

    for row in for_ratings.find_all('tr'):
        # Determine the number of rows in the table
        td_tags = row.find_all('td')
        if len(td_tags) > 0:
            n_rows+=1
            if n_columns == 0:
                # Set the number of columns for our table
                n_columns = len(td_tags)
                
        
        th_tags = for_ratings.find_all('th') 
        if len(th_tags) > 0 and len(column_names) == 0:
            for th in th_tags:
                column_names.append(th.get_text())
                
    columns = column_names if len(column_names) > 0 else range(0,n_columns)
    index=range(0,n_rows)
    df = pd.DataFrame(index=index ,columns = columns)
    
    row_marker = 0
    for row in for_ratings.find_all('tr'):
        column_marker = 0
        rows = row.find_all('td')
        for i in rows:
            df.iloc[row_marker,column_marker] = i.get_text()
            column_marker += 1
        if len(rows) > 0:
            row_marker += 1

            
    return df
    
```

```python
# import dates
start = pd.to_datetime("2019-01-12", format="%Y-%m-%d")
end = pd.to_datetime("2019-02-13", format="%Y-%m-%d")
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]

dates_to_be_crawled=[]
for date in date_generated:
    dates_to_be_crawled.append(date.strftime("%d-%m-%Y"))
```

```python
last_1_month_ratings={}
for i in range(len(dates_to_be_crawled)):
    last_1_month_ratings[i]=rating_crawler(dates_to_be_crawled[i])
```
