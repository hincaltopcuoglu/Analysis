```python
def correlation_matrix(df):
    from matplotlib import pyplot as plt
    from matplotlib import cm as cm

    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    cmap = cm.get_cmap('jet', 50)
    cax = ax1.imshow(df.corr(), interpolation="nearest", cmap=cmap)
    ax1.grid(True)
    plt.title('Feature Correlation')
    labels=['SCKLK', 'pH', 'CO', 'EC', 'OD', 'CL', 'SO4', 'NH4_N', 'NO2_N', 'NO3_N',
       'TP', 'TÇM', 'Na', 'KOI', 'TKN', 'BOI', 'YG', 'MBAS', 'Hg', 'Cd', 'Pb',
       'As', 'Cu', 'T_Cr', 'CR_6', 'COB', 'Ni', 'Zn', 'TCn', 'F', 'SCI', 'S',
       'Fe', 'Mn', 'B', 'Se', 'Ba', 'Al', 'FKOL', 'TKOL', 'Pt_Co',]
    ax1.set_xticklabels(labels,fontsize=8)
    ax1.set_yticklabels(labels,fontsize=8)
    # Add colorbar, make sure to specify tick locations to match desired ticklabels
    fig.colorbar(cax, ticks=[.75,.8,.85,.90,.95,1])
    plt.show()

correlation_matrix(df.iloc[:,12:53])
```

```python
#kuzey ve güney illerinde 
# sıcaklık bazında 2 derece , ph bazında 1 birimlik bir farklılık mevcut.Bu aradaki farklılığın anlamlı olup olmadığına bakalım.
#her bir ildeki toplanan ornekler birbirinden bağımsız olarak toplandığı için t-testi yapıp sıcaklık ve ph değerleri için
#aradaki farkın anlamlı olup olmadığına bakalım
# compare samples
from scipy.stats import ttest_ind

kuzey=df[df['İlgrup']== 'Kuzey İlleri']['SCKLK']
guney=df[df['İlgrup']== 'Güney İlleri']['SCKLK']

stat, p = ttest_ind(kuzey, guney)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distributions (fail to reject H0)')
else:
    print('Different distributions (reject H0)')
```

```python
#ph açısından bir farklılık varmı yok mu bakalım
from scipy.stats import ttest_ind

kuzey=df[df['İlgrup']== 'Kuzey İlleri']['pH']
guney=df[df['İlgrup']== 'Güney İlleri']['pH']

stat, p = ttest_ind(kuzey, guney)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distributions (fail to reject H0)')
else:
    print('Different distributions (reject H0)')
```

```python
#clustering yapabilmek için uç değerleri atmak gerekir
import numpy as np
clus = df2[['IL','İlgrup','Fe','Mn','Cu','Zn','F','B']]

xx = clus[(clus.iloc[:,2:] < clus.iloc[:,2:].quantile(.95)).all(1)]

print(xx)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

#yukarda oluşturduğum dataset i kullanıcam
#To give equal importance to all features, we need to scale the continuous features. 
#We will be using scikit-learn’s MinMaxScaler as the feature matrix is a mix of binary and continuous features
mms = MinMaxScaler()
mms.fit(xx.iloc[:,2:])
data_transformed = mms.transform(xx.iloc[:,2:])
#For each k value, we will initialise k-means 
#and use the inertia attribute to identify the sum of squared distances of samples to the nearest cluster centre
Sum_of_squared_distances = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)


plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
    
    
```

```python
# 3 clusters are optimal graph says 
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_blobs

kmeans = KMeans(n_clusters=3)
kmeans.fit(xx.iloc[:,2:])
y_kmeans = kmeans.predict(xx.iloc[:,2:])

plt.scatter(xx.iloc[:, 2], xx.iloc[:, 3], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
#2 nolu cluster demir,mangan,bakır,çinko ve bor açısından en istenmeyen maddelerin olduğu en tehlikeli cluster
#1 nolu cluster yukarıdaki özelliklere göre daha orta sınıf risk içeren cluster
#0 en risksiz cluster

#atama yapalım

xx['Cluster'] = y_kmeans

for i in range(len(xx)):
    if xx.iloc[i,8]==2:
        xx.iloc[i,8]='CokRiskli'
    elif xx.iloc[i,8]==1:
        xx.iloc[i,8]='OrtaRiskli'
    else:
        xx.iloc[i,8]='AzRiskli'
```
